```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
df=df.drop_duplicates()
```

```python
df.isna().sum()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,cmap='cool')
```

```python
##LINE CHART
```

```python
sns.lineplot(data=df,x=df.age , y=df.sex, hue='output')
```

```python
plt.plot(df['age'], df['sex'], color='blue', label="Line Plot")

plt.show()
```

```python
#hist plot
sns.histplot(data=df,x=df.trtbps, hue='sex')
```

```python
plt.hist(df['age'], bins = 20, color='blue', label="Line Plot")

plt.show()
```

```python
#scatterplot

sns.scatterplot(data=df, x=df.output,y=df.age, hue='sex')
```

```python
plt.scatter(df['age'], df['sex'], color='blue')

plt.show()
```
