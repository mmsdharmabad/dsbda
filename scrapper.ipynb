```python
request1.content
```

```python
soup= bs4.BeautifulSoup(request1.text)
```

```python
soup
```

```python
#fetching reviews and comments
```

```python
reviews = soup.findAll('div',{'class': 'ZmyHeo'});
for review in reviews:
    print(review.get_text() + "\n\n")
```

```python
#fetch rating
#avg overall rating
```

```python
rating = soup.find('div',{'class':'ipqd2A'}).get_text();
print(rating)
```

```python
#individual rating
```

```python
individual_ratings= soup.findAll('div',{'class':'XQDdHH Ga3i8K'});
for individual_rating in individual_ratings:
    print(individual_rating.get_text()+"\n\n")
```

```python
#fetching customer name
```

```python
customer_name= soup.findAll('p',{'class':'_2NsDsF AwS1CA'});
for cust_name in customer_name:
    print(cust_name.get_text()+"\n")
```

```python
#price

price = soup.find("div",{"class" : "Nx9bqj CxhGGd"})
if price:
    print(price.get_text())
else:
    print("Price not found.")
```
